```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("heart_disease_cleaned.csv")
```

```python
ds
```

```python
ds.info()
```

```python
if 'unnamed:0' in ds.columns:
    ds = ds.drop(columns=['unnamed:0'])
```

```python
#target variables distribution
plt.figure(figsize=(6,4))
sns.countplot(x='target',data = ds)
plt.title('Heart Disease Presence(Target)')
plt.xlabel('Target(0:No Disease ,1:Disease)')
plt.ylabel('Count')
plt.show()
```

```python
# age distubution
plt.figure(figsize=(6, 4))
sns.histplot(ds['age'], kde=True, bins=15)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Cholesterol Distribution

plt.figure(figsize=(6, 4))
sns.histplot(ds['chol'], kde=True, bins=20)
plt.title('Cholesterol Level Distribution')
plt.xlabel('Cholesterol (mg/dL)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Resting Blood Pressure (trestbps)

plt.figure(figsize=(6, 4))
sns.boxplot(x='target', y='trestbps', data=ds)
plt.title('Resting Blood Pressure vs Heart Disease')
plt.show()
```

```python
# Maximum Heart Rate (thalach)

plt.figure(figsize=(6, 4))
sns.boxplot(x='target', y='thalach', data=ds)
plt.title('Max Heart Rate vs Heart Disease')
plt.show()
```
